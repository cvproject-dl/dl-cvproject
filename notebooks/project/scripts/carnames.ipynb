```
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
import json
```

```
carlist = []
page = 1
cardict = {}
base_url = "https://www.india.com/auto/car-finder/price-100000-220000000/"
url = base_url
while page < 29:
  r = requests.get(url)
  soup = BeautifulSoup(r.text,"lxml")
  sec = soup.find("div",class_="car-wrap-R")

  for x in sec.find_all("div",class_='cr-list-wrap'):
    carlist.append(x.find("img")["title"])
    cardict[x.find("img")["title"]] = x.find("img")['src']

  page += 1
  url = base_url + "page-"+str(page)
  print(url)
  




  
```
